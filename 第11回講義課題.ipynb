```python
import pandas as pd
import io
#データ①「Bank_Personal_Loan_Modelling.csv」
#データを読み込む
input_data = pd.read_csv("Bank_Personal_Loan_Modelling.csv")
```

```python
input_data.head(10)
```

```python
input_data.info()
```

```python
input_data.describe()
```

```python
print(input_data['Personal Loan'])
```

```python
input_data.info()
```

```python
input_data['Personal Loan'] = pd.to_numeric(input_data['Personal Loan'], errors="coerce")
input_data.info()
```

```python
input_data.describe()
```

```python
#平均をもとめる
print(input_data['Personal Loan'].mean())
#中央値をもとめる
print(input_data['Personal Loan'].median())
#分散をもとめる
print(input_data['Personal Loan'].var())
#標準偏差をもとめる
print(input_data['Personal Loan'].std())
#最大値をもとめる
print(input_data['Personal Loan'].max())
#最小値をもとめる
print(input_data['Personal Loan'].min())
```

```python
print(input_data["Age"])
```

```python
#年齢ごとに、ローンの平均をもとめる
print(input_data.groupby('Age')['Personal Loan'].mean())
#年齢ごとに、ローンの中央値をもとめる
print(input_data.groupby('Age')['Personal Loan'].median())
#年齢ごとに、ローンの分散をもとめる
print(input_data.groupby('Age')['Personal Loan'].var())
#年齢ごとに、ローンの標準偏差をもとめる
print(input_data.groupby('Age')['Personal Loan'].std())
#年齢ごとに、ローンの最大値をもとめる
print(input_data.groupby('Age')['Personal Loan'].max())
#年齢ごとに、ローンの最小値をもとめる
print(input_data.groupby('Age')['Personal Loan'].min())
```

```python
#クレカ支払い額の平均をもとめる
print((input_data["CreditCard"]).mean())
#クレカ支払い額の中央値をもとめる
print((input_data["CreditCard"]).median())
#クレカ支払い額の分散をもとめる
print((input_data["CreditCard"]).var())
#クレカ支払い額の標準偏差をもとめる
print((input_data["CreditCard"]).std())
#クレカ支払い額の最大値をもとめる
print((input_data["CreditCard"]).max())
#クレカ支払い額の最小値をもとめる
print((input_data["CreditCard"]).min())
```

```python
input_data["Amount due"] = input_data["Personal Loan"]-input_data["CreditCard"]
```

```python
print(input_data["Personal Loan"])
print(input_data["CreditCard"])
print(input_data["Amount due"])
```

```python
import matplotlib.pyplot as plt

#ローン(Personal Loan) についてヒストグラムを描く
plt.hist(input_data['Personal Loan'])
plt.xlabel('Personal Loan')
plt.ylabel('count')
plt.grid(True)
```

```python
import matplotlib.pyplot as plt
import scipy.stats

plt.scatter(input_data["Age"],input_data['Personal Loan'], s=50, c="blue", alpha=0.3, linewidths=1, edgecolors="blue")
plt.title('Age and Personal Loan')
plt.ylabel('Personal Loan')
plt.xlabel('Age')
```

```python
import seaborn as sns
sns.jointplot(data=input_data, x="Age", y="Personal Loan")
```

```python
import numpy as np

#モデルで利用する変数を接合し, ひとつの DataFrame にする
#読み込むデータを変えれば, 説明変数や被説明変数を変えることが可能
x_list = pd.concat([input_data['Personal Loan'], input_data["CCAvg"],input_data["Mortgage"], input_data["Online"], input_data["Income"]], axis=1)

#pairplotをプロットする
sns.pairplot(x_list, kind='reg')
```
