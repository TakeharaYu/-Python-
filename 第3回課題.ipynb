```python
#手書き文字のデータ(load_digits)を読み込む
from sklearn import datasets
dataset = datasets.load_digits()
dataset
```

```python
#pandas パッケージを利用する
import pandas as pd
#DataFrame に代入する
dataset2=pd.DataFrame(dataset.data)
#列名を指定する
dataset2.columns=dataset.feature_names
#結果を出力する
dataset2
```

```python
#sklearn からdataset を利用する
from sklearn import datasets
#pandas パッケージを利用する
import pandas as pd
#load_linnerudでデータセットのサンプルを読み込む
dataset = datasets.load_linnerud()
#DataFrame に代入する
dataset2=pd.DataFrame(dataset.data)
#列名を指定する
dataset2.columns=dataset.feature_names
#種類を取り出す
dataset3=pd.DataFrame(dataset.target)
#列名を任意に指定する
dataset3.columns = dataset.target_names
#ふたつのデータセットを縦につなぎ合わせる
dataset4=pd.concat([dataset3, dataset2], axis=1)
#結果を出力する
dataset4
```

```python
#sklearn からdataset を利用する
from sklearn import datasets
#pandas パッケージを利用する
import pandas as pd
#load_digitsでデータセットのサンプルを読み込む
dataset = datasets.load_digits()
#DataFrame に代入する
dataset2=pd.DataFrame(dataset.data)
#列名を指定する
dataset2.columns=dataset.feature_names
#クラスを取り出す
dataset3=pd.DataFrame(dataset.target)
#列名を任意に指定する
dataset3.columns=['ID']
#ふたつのデータセットを縦につなぎ合わせる
dataset4=pd.concat([dataset3, dataset2], axis=1)
#結果を出力する
dataset4
```
